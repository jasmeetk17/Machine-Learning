# 3 Opening a csv file from Url

```python
import requests
from io import StringIO

url="https://gist.github.com/rnirmal/e01acfdaf54a6f9b24e91ba4cae63518"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"} # Changed headers to a dictionary
req=requests.get(url,headers=headers)

data=StringIO(req.text)
pd.read_csv(data,sep='\\t')
```
